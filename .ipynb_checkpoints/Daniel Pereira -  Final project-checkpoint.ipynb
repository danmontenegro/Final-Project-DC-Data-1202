```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sv_raw = pd.read_csv("survey.csv")
sv_raw.head()
```

```python
sv_treat = sv_raw[sv_raw["treatment"] == "Yes"]
sv_treat.head()
```

```python
##################################
#####Create a Function ###########
##################################
def sv_group(df, Sort_Value, agr_col):

    """ This function creates a groupby by a specific column and returns the sorted df

    df = a dataframe you want to analyse 

    Agr_value = The column that you want to aggregate"""

    sv_count_df = df.groupby(Sort_Value)[agr_col].count().reset_index()
    sv_count_df = sv_count_df.sort_values(agr_col, ascending=False)
    return sv_count_df
```

```python
##People in treatment per country

sv_count_treat = sv_group(sv_treat,"Country", "treatment")
print(sv_count_treat)
```

```python
fig, ax = plt.subplots()
ax.bar(x=sv_count_treat.Country,height=sv_count_treat.treatment)
plt.title('Country Vs Treatment')
plt.xticks(rotation='vertical')
plt.xlabel('Country')
plt.ylabel('People in Treatment')
plt.tight_layout(h_pad=5)
fig.savefig('Ppl_in_treat_per_count.png')
```

```python
#New dataset with Companyes with benefits
sv_benef = sv_raw[sv_raw["benefits"] == "Yes"]
sv_benef.head()
```

```python
##  Top 10 Countries - Benefits 

sv_count_benefits = sv_group(sv_benef,"Country", "benefits").sort_values("Country", ascending=False).reset_index()
sv_top10_ben = sv_count_benefits.iloc[0:10]
print(sv_top10_ben)
```

```python
###Plot Top 10  Countries Benefits

fig, ax = plt.subplots()
ax.bar(x=sv_top10_ben.Country,height=sv_top10_ben.benefits)
plt.title('Companies with Benefits per country')
plt.xticks(rotation='vertical')
plt.xlabel('Country')
plt.ylabel('People who have benefits inside company')
plt.tight_layout(h_pad=5)
fig.savefig('Top_10_Benefits.png')
```

```python
###   Companies interviewed per country 

sv_count_companies = sv_group(sv_raw,"Country", "benefits").sort_values("benefits", ascending=False)
sv_top10_comp = sv_count_companies.iloc[0:10]
print(sv_top10_comp)
```

```python
# Plot Companies interviewd

fig, ax = plt.subplots()
ax.bar(x=sv_top10_comp.Country,height=sv_top10_comp.benefits)
plt.title('Companies per country')
plt.xticks(rotation='vertical')
plt.xlabel('Country')
plt.ylabel('People interviewed')
plt.tight_layout(h_pad=10)
fig.savefig('Companies Interviewd.png')
```

```python
# Total of Companies that have at least one employee with benefits
sv_total = sv_count_companies[sv_count_companies["Country"].isin(sv_count_benefits["Country"])].sort_values("Country", ascending=False).reset_index()
print(sv_total)
```

```python
### Percentage of people with benefits per country

sv_count_benefits["Percentage"] = sv_count_benefits["benefits"] / sv_total["benefits"]

sv_count_benefits = sv_count_benefits.sort_values("Percentage", ascending=False)
print(sv_count_benefits)
```

```python
### Plot of Percentage of people with benefits per country 
fig, ax = plt.subplots()
ax.bar(x=sv_count_benefits.Country, height=sv_count_benefits.Percentage)
plt.title('Companies per country')
plt.xticks(rotation='vertical')
plt.xlabel('Country')
plt.ylabel('People interviewed')
plt.tight_layout(h_pad=10)
```
